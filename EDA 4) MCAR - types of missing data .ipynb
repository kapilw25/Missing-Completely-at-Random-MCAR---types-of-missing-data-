# Types of Missing Data:
# 1) Missing Completely at Random (MCAR):

A variable is missing completely at random (MCAR)if the missing values on a given variable (Y) don’t have a relationship with other variables in a given data set or with the variable (Y) itself. In other words, 
##### When data is MCAR, there is no relationship between the data missing and any values, and there is no particular reason for the missing values.
